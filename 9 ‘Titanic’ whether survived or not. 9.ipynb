```python
import seaborn as sns 
```

```python
df = sns.load_dataset('titanic') 
```

```python
df
```

```python
sns.boxplot(x='sex',y='age', data=df)
```

```python
sns.boxplot(x='sex',y='age', hue='survived', data=df) 
```

```python
sns.scatterplot(x='sex', y='age', hue='survived', data=df)
```

```python
import matplotlib.pyplot as plt
plt.title('Distribution of Age with respect to Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
sns.countplot(df['survived'])
```

```python
df['sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
plt.hist(df['age'], bins=5)
plt.show()
```

```python
sns.distplot(df['age'])
plt.show()
```

```python
sns.barplot(df['pclass'], df['age'])
plt.show()
```

```python
sns.distplot(df[df['survived'] == 0]['age'], hist=False, color="blue")
sns.distplot(df[df['survived'] == 1]['age'], hist=False, color="orange")
```

```python
import pandas as pd
pd.crosstab(df['pclass'], df['survived'])
```

```python
sns.heatmap(pd.crosstab(df['pclass'], df['survived']))
```

```python
sns.clustermap(pd.crosstab(df['parch'], df['survived']))
plt.show()
```
